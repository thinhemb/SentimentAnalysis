```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("/home/thinh_do/Workplace/SentimentAnalysis/data.csv")
# df["Average_training_loss"]
```

```python
def plot_loss():
    train_losses = df['Loss_train']
    at_losses = df['Average_training_loss']
    plt.plot(train_losses,marker = 'o',label='Training loss')
    plt.plot(at_losses,label='Average_training_loss',marker = 'o')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(' Loss')
    plt.legend()
    plt.savefig('./Loss.png', dpi=300)
plot_loss()
```

```python
def plot_acc():
    train_acc = df['Accuracy_train']
    val_acc = df['Accuracy_val']
    plt.plot(train_acc,marker = 'o',label='Accuracy_train')
    plt.plot(val_acc,label='Accuracy_val',marker = 'o')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(' Accuracy')
    plt.legend()
    plt.savefig('./Accuracy.png', dpi=300)
plot_acc()
```

```python
def plot_F1():
    train_acc = df['F1_score_train']
    val_acc = df['F1_score_val']
    plt.plot(train_acc,marker = 'o',label='F1_score_train')
    plt.plot(val_acc,label='F1_score_val',marker = 'o')
    plt.xlabel('Epoch')
    plt.ylabel('F1 score')
    plt.title('F1 score')
    plt.legend()
    plt.savefig('./F1_score.png', dpi=300)
plot_F1()
```
